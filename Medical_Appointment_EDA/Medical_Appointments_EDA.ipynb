```python
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay']).dt.date.astype('datetime64[ns]')
```

```python
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay']).dt.date.astype('datetime64[ns]')
```

```python
df['Sch_WOD'] = df['ScheduledDay'].dt.dayofweek
```

```python
df['App_DOW'] = df['AppointmentDay'].dt.dayofweek
```

```python
df
```

```python
df['Sch_WOD'].value_counts()
```

```python
df['Sch_WOD'] = df['Sch_WOD'].map({0:'Monday',
                  1:'Tuesday',
                  2:'Wednesday',
                  3:'Thursday',
                  4:'Friday',
                  5:'Saturday'})
```

```python
df['App_DOW'].value_counts()
```

```python
df['App_DOW'] = df['App_DOW'].map({0:'Monday',
                  1:'Tuesday',
                  2:'Wednesday',
                  3:'Thursday',
                  4:'Friday',
                  5:'Saturday'})
```

```python
df.head()
```

```python
df['Sch_WOD'].value_counts().plot(kind = 'bar', color = "b")
plt.xticks(rotation = -45)
plt.title('No Of Patients Schedule day')
plt.show()
```

```python
df['App_DOW'].value_counts().plot(kind = 'bar')
plt.xticks(rotation = -45)
plt.title('No Of Patients Appointment day')
plt.show()
```
