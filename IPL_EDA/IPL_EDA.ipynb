```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the ipl matches dataset
ipl = pd.read_csv(r"C:\Users\areeb\Desktop\Data Science\Data Sets\matches.csv")
```

```python
#first five record of the dataset
ipl.head()
```

```python
#looking at the number of rows and columns in the dataset
ipl.shape
```

```python
# getting the frequency of most man of the match award
ipl["player_of_match"].value_counts()
```

```python
# getting the top 5 player with most man of the match
ipl["player_of_match"].value_counts()[0:5]
```

```python
# making the bar plot for the top 5 players with most of the match awards
plt.figure(figsize = (8,5))
plt.bar(list(ipl["player_of_match"].value_counts()[0:5].keys()),
        list(ipl["player_of_match"].value_counts()[0:5]),
        color = "g",
        edgecolor = "black",
        linewidth = 2)
plt.show()
```

```python
# getting the frequency of result column
ipl["result"].value_counts()
```

```python
# getting the frequency of result column
ipl["toss_winner"].value_counts()
```

```python
# extracting the record where a team won batting first
batting_first = ipl[ipl["win_by_runs"]!=0]
```

```python
# Looking at the head
batting_first.head()
```

```python
# making a histogram
plt.figure(figsize=(7,7))

plt.xlabel("Runs")

plt.hist(batting_first["win_by_runs"], 
         color = "r",
         alpha = 1,
         edgecolor = "black",
         linewidth = 2)

plt.title("Distribution of Runs")
plt.grid()
plt.show()
```

```python
#finding out the number of wins w.r.t each team after batting first
batting_first["winner"].value_counts()
```

```python
# making a bar plot for top 3 teams with most wins after batting first
plt.figure(figsize=(7,7))
plt.bar(
        list(batting_first["winner"].value_counts()[0:3].keys()),
        list(batting_first["winner"].value_counts()[0:3]),
        color = ["blue","yellow","red"],
        edgecolor = "black",
        linewidth = 2
        )

plt.title("Top 3 teams with most wins after batting first")
plt.show()
```

```python
# Making a pie chart 
plt.figure(figsize = (7,7))

plt.pie(
        list(batting_first["winner"].value_counts()),
        labels = list(batting_first["winner"].value_counts().keys()),
        autopct = "%0.1f%%",
        wedgeprops = {"linewidth":2, "edgecolor":"black"},
        textprops = {"fontsize": 15}
)

plt.show()
```

```python
# Extracting the records where a team has won after batting second
batting_second = ipl[ipl["win_by_wickets"] != 0]
```

```python
#looking at the head
batting_second.head()
```

```python
# Making the histogram for frequency of w.r.t number of wickets 
plt.figure(figsize=(7,7))


plt.hist(batting_second["win_by_wickets"],
         bins = 30,
         color = "red",
         alpha = 1,
         edgecolor = "black",
         linewidth = 2)

plt.xlabel("Wickets")
plt.grid()
plt.title("Frequency of w.r.t number of wickets")
plt.show()
```

```python
# finding out the frequency of numbers wins w.r.t each time after batting second
batting_second["winner"].value_counts()
```

```python
#Making the bar plot for the top 3 with most wins after batting second
plt.figure(figsize = (7,7))

plt.bar(batting_second["winner"].value_counts().keys()[0:3],
        batting_second["winner"].value_counts()[0:3],
        color = ["purple","blue","red"],
        edgecolor = "black",
        linewidth = 2)


plt.title( "Top 3 with most wins after batting second" )
plt.show()
```

```python
# Making a pie chart for distribution of wins after second
plt.figure(figsize=(7,7))

plt.pie(batting_second["winner"].value_counts(),
        labels = batting_second["winner"].value_counts().keys(), 
        autopct = "%1.1f%%",
        wedgeprops = {"linewidth":1,"edgecolor":"black"}
       )

plt.show()
```
